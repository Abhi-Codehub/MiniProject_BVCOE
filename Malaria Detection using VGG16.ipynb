# Malaria Detection Using VGG16


```python
import tensorflow as tf
import tensorflow_datasets as tfds
import pandas as pd
import numpy as np

# Load the malaria dataset from TFDS
data, info = tfds.load("malaria", with_info=True, as_supervised=True)

# Define the pre-processing function
def preprocess(image, label):
    # Convert the image to float32 type
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)
    # Resize the image to (224, 224)
    image = tf.image.resize(image, (224, 224))
    # Return the preprocessed image and its label
    return image, label

# Preprocess the train and test splits
train_data = data['train'].map(preprocess)

# Define the VGG16 model
vgg16 = tf.keras.applications.VGG16(include_top=False, weights='imagenet')

# Extract features from the train split using the VGG16 model
train_features = []
train_labels = []
for image, label in train_data:
    features = vgg16.predict(image[tf.newaxis, ...])[0]
    train_features.append(features)
    train_labels.append(label.numpy())
train_features = np.array(train_features)
train_labels = np.array(train_labels)

# Save the features and labels as CSV files
pd.DataFrame(train_features).to_csv('train_features.csv', index=False)
pd.DataFrame(train_labels).to_csv('train_labels.csv', index=False)

from google.colab import files

# Download the CSV file to your computer
files.download('file_name.csv')

import tensorflow_datasets as tfds

data, info = tfds.load("malaria", with_info=True, as_supervised=True)
print(info.splits)

from google.colab import files

# Download the CSV file to your computer
files.download('train_features.csv')
```
